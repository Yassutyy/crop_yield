### classification model (High Yield vs Low Yield).

### Create High vs Low Yield Column

We’ll use the median yield as the threshold (above = High Yield, below = Low Yield).

```python
# Create classification target
median_yield = df['Yield'].median()
df['Yield_Class'] = (df['Yield'] >= median_yield).astype(int)  

# 1 = High Yield, 0 = Low Yield
df[['Yield', 'Yield_Class']].head()
```

### Define Features (X) and New Target (y_class ) 

```python
X = df[['Fertilizer', 'Temp', 'N', 'P', 'K']]
y_class = df['Yield_Class']

# Train/test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train_class, y_test_class = train_test_split(
    X, y_class, test_size=0.2, random_state=42
)
```

### Logistic Regression Classifier

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Train
log_clf = LogisticRegression(max_iter=1000)
log_clf.fit(X_train, y_train_class)

# Predictions
y_pred_log = log_clf.predict(X_test)

# Evaluation
print("📊 Logistic Regression Classifier:")
print("Accuracy:", accuracy_score(y_test_class, y_pred_log))
print("\nClassification Report:\n", classification_report(y_test_class, y_pred_log))
print("\nConfusion Matrix:\n", confusion_matrix(y_test_class, y_pred_log))
```

### Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier

# Train
dt_clf = DecisionTreeClassifier(max_depth=5, random_state=42)
dt_clf.fit(X_train, y_train_class)

# Predictions
y_pred_dt = dt_clf.predict(X_test)

# Evaluation
print("📊 Decision Tree Classifier:")
print("Accuracy:", accuracy_score(y_test_class, y_pred_dt))
print("\nClassification Report:\n", classification_report(y_test_class, y_pred_dt))
print("\nConfusion Matrix:\n", confusion_matrix(y_test_class, y_pred_dt))
```

Now let’s visualize the classification results so you can clearly see model performance.

### Confusion Matrix Heatmap (for Decision Tree as example)

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

# Confusion matrix for Decision Tree
cm = confusion_matrix(y_test_class, y_pred_dt)

plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=["Low Yield", "High Yield"],
            yticklabels=["Low Yield", "High Yield"])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Decision Tree - Confusion Matrix")
plt.show()
```
